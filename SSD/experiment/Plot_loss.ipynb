```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open('stats_RMSProp.pkl','rb') as f:
    l_loss, c_loss, itr = pickle.load(f)
```

```python
l_loss = np.asarray(l_loss)
c_loss = np.asarray(c_loss)
itr = np.asarray(itr)
```

```python
fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes

ax.set_xlabel('Iterations', fontsize=14)
ax.set_ylabel('SSD Total Training Loss', fontsize=14)

ax.set_title('SSD Training Loss', fontsize=15)
loss = ax.plot(itr[7:700],l_loss[7:700]+c_loss[7:700])

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('ssd_training_loss.png', bbox_inches='tight')
```

```python
fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes

ax.set_xlabel('Iterations', fontsize=14)
ax.set_ylabel('SSD Confidence Loss', fontsize=14)

ax.set_title('SSD Training Loss', fontsize=15)
loss_c = ax.plot(itr[7:700],c_loss[7:700])

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('ssd_training_confidence_loss.png', bbox_inches='tight')
```
